"In India, crop production plays a vital role in the country's agricultural sector. With a diverse range of agro-climatic zones, India cultivates a wide variety of crops throughout the year. Major crops include rice, wheat, maize, pulses, oilseeds, cotton, sugarcane, fruits, and vegetables. The production of crops in India is influenced by factors such as monsoon rainfall, irrigation systems, technological advancements, government policies, and market demand. Crop production practices vary across different states and regions, with farmers adopting traditional methods as well as modern techniques like precision agriculture. The agricultural sector contributes significantly to India's economy and provides livelihoods to millions of farmers across the country."

```python
areas_of_production =df.groupby(['State'])['Cost of Production (`/Quintal) C2'].max().reset_index()
areas_of_production
```

```python
area = areas_of_production['State']
price = areas_of_production['Cost of Production (`/Quintal) C2']

fig = plt.figure(figsize = (19, 7))

# creating the bar plot
plt.bar(area, price, color ='magenta',
        width = 0.4)

plt.xlabel("STATE")
plt.ylabel("PRICE")
plt.title("Cost of Production  vs  State")
plt.show()
```

```python
areas_of_production =df.groupby(['State'])['Crop'].max().reset_index()
areas_of_production
```

```python
crop= areas_of_production['State']
region = areas_of_production['Crop']

fig = plt.figure(figsize = (19, 4))

# creating the bar plot
plt.plot(crop, region)

plt.xlabel("REGION")
plt.ylabel("CROP")
plt.title("CROP  vs  REGION")
plt.show()
```
